### Выявление аномалий в стоимости автомобиля

```python
df = pd.read_csv('data/vehicles_dataset_for_dv.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.hist(df.price, bins=1000, color='indigo')
ax.set_xlabel('Стоимость автомобиля')
ax.set_ylabel('Частота')
ax.set_title('Распределение стоимости автомобиля')

ax.xaxis.set_major_formatter(FormatStrFormatter('%.0f'));
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

ax.boxplot(df.price, vert=False)
ax.set_xlabel('Стоимость автомобиля')
ax.set_title('Распределение стоимости автомобиля')

ax.xaxis.set_major_formatter(FormatStrFormatter('%.0f'));
```

Jписательные статистики стоимости автомобиля, чтобы изучить значения, обозначающие границы «ящика». 

```python
df.price.describe().apply(lambda x: f"{x:0.3f}")
```

Границы для нормальных значений стоимости автомобиля

```python
q25 = df.price.quantile(0.25)
q75 = df.price.quantile(0.75)
iqr = q75 - q25
boundaries = (q25 - 1.5 * iqr, q75 + 1.5 * iqr)
boundaries
```

Количество выбросов в стоимости автомобиля. 

```python
is_outlier = (df.price < boundaries[0]) | (df.price > boundaries[1])
is_outlier.sum()
```

Записи, в которых стоимость автомобиля попадает в выбросы, отсортированные по стоимости

```python
df_outlier_sorted_by_price = df[is_outlier].sort_values('price')
df_outlier_sorted_by_price
```

Последние 15 записей в отсортированном ранее датафрейме с выбросами, чтобы изучить признаки для наиболее дорогих автомобилей.

```python
df_outlier_sorted_by_price.tail(15)
```

Среди автомобилей с аномальной стоимостью действительно присуствуют автомобили, высокую стоимость которых можно объяснить, в первучю очередь, производителем и низким пробегом. Но будем опираться на стоимость основной массы автомобилей, поэтому удалим все аномальные значения и сохраним датафрейм без аномалий в переменную `df_new`

```python
df_new = df.loc[df['price'] < 1000000]
df_new.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

ax.boxplot(df_new.price, vert=False)
ax.set_xlabel('Стоимость автомобиля')
ax.set_title('Распределение стоимости автомобиля')

ax.xaxis.set_major_formatter(FormatStrFormatter('%.0f'));
```

### Исследование зависимости стоимости автомобиля от года производства

Датафрейм, в котором будет содержаться статистика о годе производства автомобиля и среднем значении стоимости автомобиля за этот год

```python
stats_year = df_new.groupby('year', as_index=False)['price'].mean()
stats_year = stats_year.rename(columns={'price': 'price_mean'})
stats_year
```

```python
plt.figure()
plt.scatter(stats_year['price_mean'], stats_year['year'])
plt.title("Зависимость стоимости от года выпуска автомобиля")
plt.xlabel('Стоимость')
plt.ylabel('Год выпуска')

plt.show();
```

Из графика можно сделать вывод о возрастающем тренде: чем новее автомобиль, тем выше его цена. Исключение составляют наиболее старые и самые новые автомобили. Проверим, достаточно ли данных в выбивающихся из тренда значениях по годам. Для этого выведем количество автомобилей, произведенных в каждый год

```python
df_new.groupby('year')['id'].count()
```

За последние два года мало накопленной информации — исключим их из исследования. С помощью фильтрации оставим в датафрейме автомобили, произведенные не позднее 2021 года

```python
df_new2 = df_new.loc[df['year'] < 2021]
df_new2.head()
```

На отфильтрованных данных повторим действия по формированию нового датафрейма, в котором будет содержаться статистика о годе производства автомобиля и среднем значении стоимости автомобиля за этот год

```python
stats_year_quantity = df_new2.groupby('year', as_index=False)['id'].count()
stats_year_quantity = stats_year_quantity.rename(columns={'id': 'count'})
stats_year_quantity
```

```python
plt.figure(figsize=(8, 4))
plt.plot(stats_year_quantity['year'], stats_year_quantity['count'], color='blue')
plt.xlabel('Год выпуска автомобиля')
plt.ylabel('Количество автомобилей')
plt.title('Зависимость количества автомобилей от его года выпуска');
```

Посчитаем коэффициент корреляции Пирсона для стоимости автомобиля и года его производства.

```python
stats_year_quantity.corr()
```

Для обновленных данных построим гистограмму распределения стоимости автомобиля

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.hist(df_new.price, bins=20, color='indigo')
ax.set_xlabel('Стоимость автомобиля')
ax.set_ylabel('Частота')
ax.set_title('Распределение стоимости автомобиля')

ax.xaxis.set_major_formatter(FormatStrFormatter('%.0f'));
```

### Исследование зависимости ценовой категории автомобилиля от стоимости и года производства

```python
colors = {'low': 'green', 'medium': 'orange', 'high': 'red'}

fig, ax = plt.subplots(figsize=(12, 8))
for price_category in colors:
    color = colors[price_category]
    data = df_new[df_new['price_category'] == price_category]

    ax.hist(data['price'], bins=20, color=color, alpha=0.7)
    
plt.ylabel('Частота')
plt.xlabel('Стоимость автомобиля')
plt.title('Распределение стоимости автомобиля для каждой из ценовой категории автомобил')
plt.legend(colors, title="Ценовая категория");
```

```python
colors = {'low': 'green', 'medium': 'orange', 'high': 'red'}
fig, ax = plt.subplots(figsize=(12, 8))
for price_category in colors:
    color = colors[price_category]
    data = df_new[df_new['price_category'] == price_category]

    ax.scatter(data['price'], data['year'], c=color, label=price_category)

ax.legend(loc='upper right', title="Ценовая категория")
plt.title("Зависимость стоимости от года выпуска автомобиля")
plt.xlabel('Стоимость')
plt.ylabel('Год выпуска')
plt.show()
```

### Визуализация категориальных переменных

Датафрейм, в котором будет содержаться статистика о производителе автомобиля и количестве произведенных им автомобилей 

```python
stats_manufacturer = df.groupby('manufacturer', as_index=False)['id'].count()
stats_manufacturer = stats_manufacturer.rename(columns={'id': 'count'})
stats_manufacturer
```

```python
plt.figure(figsize=(20, 10))
plt.bar(stats_manufacturer['manufacturer'], stats_manufacturer['count'], width=0.9, color=['springgreen','dodgerblue'])
plt.xticks(rotation=45)
plt.xlabel('Призводитель автомобиля')
plt.ylabel('Количество автомобилей')
plt.title('Зависимость количества автомобилей от производителя');
```

Датафрейм, в котором будет содержаться статистика о коробке передач автомобиля и количестве автомобилей с такой коробкой передач

```python
stats_transmission = df.groupby('transmission', as_index=False)['id'].count()
stats_transmission = stats_transmission.rename(columns={'id': 'count'})
stats_transmission
```

```python
plt.figure(figsize=(12, 8))
plt.pie(stats_transmission['count'], labels=stats_transmission['transmission'], 
        labeldistance=None, autopct='%1.0f%%', colors=['aquamarine','salmon','wheat'])
plt.title("Распределение коробок передач")
plt.legend()
plt.show();
```
